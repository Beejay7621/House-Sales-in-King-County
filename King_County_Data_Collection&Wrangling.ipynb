```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline 
```

```python
file_name='https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DA0101EN-SkillsNetwork/labs/FinalModule_Coursera/data/kc_house_data_NaN.csv'
df=pd.read_csv(file_name)
```

```python
df.head()
```

```python
# Checking for columns with missing values.  From the output, we can tell that "bedrooms" and "bathrooms"
# have missing values
df.info()
```

```python
features = ['price', 'bedrooms', 'bathrooms', 'floors', 'lat', 'long']
df[features] = df[features].astype('float32')
df.info()
```

```python
# Checking for total number of missing values
print(df["bedrooms"].isnull().sum())
print(df["bathrooms"].isnull().sum())
```

```python
# Defining the mean of the column "bedrooms".
bedrooms_mean = df["bedrooms"].mean()
```

```python
# Replacing the missing values with the mean
df["bedrooms"].replace(np.nan, bedrooms_mean, inplace = True)
```

```python
# Defining the mean of the column "bathrooms"
bathrooms_mean = df["bathrooms"].mean()
```

```python
# Replacing the missing values with the mean
df["bathrooms"].replace(np.nan, bathrooms_mean, inplace = True)
```

```python
# Confirming the missing values have been replaced with the mean of the corresponding columns
print(df["bedrooms"].isnull().sum())
print(df["bathrooms"].isnull().sum())
```

```python
# The columns "id" and "Unnamed: 0" are redundant.  These columns need to be dropped
df.drop(["id", "Unnamed: 0"], axis = 1, inplace = True)
```
